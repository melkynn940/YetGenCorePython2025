# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_acik = False
        self.ses = 10
        self.kanallar = ["TRT", "Kanal D", "Fox"]
        self.aktif_kanal = self.kanallar[0]

    def tv_ac(self):
        if not self.tv_acik:
            self.tv_acik = True
            print("Televizyon açıldı.")
        else:
            print("Televizyon zaten açık.")

    def tv_kapat(self):
        if self.tv_acik:
            self.tv_acik = False
            print("Televizyon kapandı.")
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayar(self):
        print(f"Mevcut ses seviyesi: {self.ses}")
        try:
            yeni_ses = int(input("Yeni ses seviyesi (0-100): "))
            if 0 <= yeni_ses <= 100:
                self.ses = yeni_ses
                print(f"Ses seviyesi {self.ses} olarak ayarlandı.")
            else:
                print("Geçersiz seviye.")
        except ValueError:
            print("Lütfen sayı girin.")

    def kanal_ekle(self):
        kanal = input("Eklenecek kanal adı: ")
        self.kanallar.append(kanal)
        print(f"{kanal} kanallara eklendi.")

    def kanal_sayisi(self):
        print(f"Toplam kanal sayısı: {len(self.kanallar)}")

    def rastgele_kanal(self):
        self.aktif_kanal = random.choice(self.kanallar)
        print(f"Şu anki kanal: {self.aktif_kanal}")

    def tv_bilgileri(self):
        print(f"Tv durumu: {'Açık' if self.tv_acik else 'Kapalı'}")
        print(f"Ses seviyesi: {self.ses}")
        print(f"Aktif kanal: {self.aktif_kanal}")
        print(f"Kanallar: {self.kanallar}")

kumanda = Kumanda()

while True:
    secim = input("""
1 - Tv Aç
2 - Tv Kapat
3 - Ses Ayarları
4 - Kanal Ekle
5 - Kanal Sayısını Öğren
6 - Rastgele Kanala Geç
7 - Televizyon Bilgileri
q - Çıkış
Seçiminiz: """)

    if secim == "1":
        kumanda.tv_ac()
    elif secim == "2":
        kumanda.tv_kapat()
    elif secim == "3":
        kumanda.ses_ayar()
    elif secim == "4":
        kumanda.kanal_ekle()
    elif secim == "5":
        kumanda.kanal_sayisi()
    elif secim == "6":
        kumanda.rastgele_kanal()
    elif secim == "7":
        kumanda.tv_bilgileri()
    elif secim == "q":
        print("Çıkış yapılıyor...")
        break
    else:
        print("Geçersiz giriş.")
```
